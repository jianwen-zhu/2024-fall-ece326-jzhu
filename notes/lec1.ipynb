# PL Paradigms


## Imperiative


* model of Von Nuemann machine
   - Store (memory)
   - ALU
   - Load/Store into memory

* Language constructs
  - variable: states (store)
  - statement: state transitions
  - expressions: computation

* Examples

Fortran
```
    SUM = 0
    DO 11 K=1,N
    SUM = SUM + 2*K
11  CONTINUE
```

C
```
  sum = 0;
  for( k = 1; k <= n; k ++ ) {
      sum += 2*k;
```

Pascal
```
  sum := 0;
  for k := 1 to n do
      sum := sum + 2*k;
```


## Functional

* Model of Dataflow machine
  - No store
* Language constructs
  - No statement 
  - iteration through recursion
  - Function application

Scheme
```
	(define (sum n)
			(if (= n 0)
				0
				(+ (* n 2) (sum (- n 1)))
			)
	)
```
